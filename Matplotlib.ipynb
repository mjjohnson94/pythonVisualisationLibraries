```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np 
```

```python
## Basic Line Chart 

# Initialise the figure of the plot & the axes 
fix, ax = plt.subplots()

# Assign x & y variables with data, this represents the X & Y axis 
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]

# Plot the data on the axes to create the visualisation, referencing the x & y variables  
ax.plot(x,y)
```

```python
## Basic Line Chart - Design Elements 

# Initialise the figure of the plot & the axes 
fix, ax = plt.subplots()

# Assign x & y variables with data, this represents the X & Y axis 
x = [1, 2, 3, 4]
y = [2, 4, 6, 8]
z = [5, 15, 10, 7]

# Plot the data on the axes to create the visualisation, referencing the x & y variables. 
# NOTE - each variable must be assigned a label to use the ax.legend() feature 
ax.plot(x, label="X", color="red", linestyle="--")
ax.plot(y, label="Y", color="darkblue", linewidth="1")
ax.plot(z, label="Z", color="black")

# Setting label for X Axis
ax.set_xlabel('This is a X Axis Label') 

# Setting label for Y Axis 
ax.set_ylabel('This is a Y Axis Label') 

# Setting title for visualisation 
ax.set_title("This is a Title")  

# Add gridlines 
ax.grid(True)

# Setting legend 
ax.legend();  
```

```python
## Basic Bar Chart

# Initialise the figure of the plot & the axes 
fix, ax = plt.subplots()

# Assign x & y variables with data, this represents the X & Y axis 
x = ["Blue", "Green", "Orange", "Yellow"]
y = [2, 4, 6, 8]

# Plot the data on the axes to create the visualisation, referencing the x & y variables and using ax.BAR
ax.bar(x, y)
```

```python
## Basic Stacked Bar Chart 


# Create list to store the X Axis tick labels 

labels = ['G1', 'G2', 'G3', 'G4', 'G5']

# Create list to store the data (integer) which will be plotted

men_means = [20, 35, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]


# Setting the width of each bar on the plot 

width = 0.35       

# Initialising the figure of the plot and the axis 

fig, ax = plt.subplots()

# Plot the data on the axis and assign label for each data set 

ax.bar(labels, men_means, width, label='Men')
ax.bar(labels, women_means, width, bottom=men_means, label='Women')


# Setting label for the Y Axis 

ax.set_ylabel('Scores')

# Setting Chart title 

ax.set_title('Scores by group and gender')

# Adding legend 

ax.legend()

# Displaying the plot 

plt.show()
```
